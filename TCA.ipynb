#convert value 0 and 1 value of senior citizen to yes/no to make it easier

```python
def conv(value):
    if value==1:
        return "yes"
    else:
         return "no"

df["Senior Citizen"] = df["Senior Citizen"].apply(conv)
```

```python
# Create the countplot
plt.figure(figsize=(6,6))
ax = sns.countplot(x="Senior Citizen", data=df, hue="Churn Label")

# Calculate the total count for each Senior Citizen category
total = len(df)

# Loop through the bars to add percentage labels
for p in ax.patches:
    height = p.get_height()  # Get the height of the bar (count)
    width = p.get_width()    # Get the width of the bar
    x = p.get_x() + width / 2  # Get the x position to place the label
    y = height  # Y position is the height of the bar
    
    # Calculate percentage
    percentage = (height / total) * 100
    
    # Display percentage label above the bar
    ax.text(x, y, f'{percentage:.1f}%', ha="center", va="bottom", fontsize=12)

# Add title
plt.title("Churn by Senior Citizen (%)", fontsize=20)
```

```python
# Assume your data is loaded into a Pandas DataFrame called df
# Example: df = pd.read_csv('your_data.csv')

# List of columns for which count plots are needed
columns = [
    'Phone Service', 'Avg Monthly Long Distance Charges', 'Multiple Lines',
    'Internet Service', 'Internet Type',
    'Online Security', 'Online Backup',
     'Streaming TV', 'Streaming Movies'
]

# Set up the grid for subplots
n_cols = 3  # Number of columns in the subplot grid
n_rows = -(-len(columns) // n_cols)  # Calculate rows needed (ceil division)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through each column and create a count plot
for i, column in enumerate(columns):
    if column in df.columns:
        sns.countplot(data=df, x=column, ax=axes[i], hue = "Churn Label")
        axes[i].set_title(column)
        axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels if needed
    else:
        axes[i].axis('off')  # Turn off unused subplot if the column doesn't exist

# Turn off any extra subplots if the number of columns is less than n_rows * n_cols
for j in range(len(columns), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
